```python
import chat_assistant
from openai import OpenAI
import random

# --- 1. Define your functions and known data (from homework) ---
known_weather_data = {
    'berlin': 20.0
}

def get_weather(city: str) -> float:
    """Retrieves the temperature for a specified city."""
    city = city.strip().lower()
    if city in known_weather_data:
        return known_weather_data[city]
    return round(random.uniform(-5, 35), 1)

def set_weather(city: str, temp: float) -> str:
    """Sets the temperature for a specified city."""
    city = city.strip().lower()
    known_weather_data[city] = temp
    return 'OK'

# --- 2. Define the tool descriptions (your answers to Q1 & Q2) ---
get_weather_tool = {
    "type": "function", "name": "get_weather", "description": "Retrieves the temperature for a specified city.",
    "parameters": {"type": "object", "properties": {"city": {"type": "string", "description": "The name of the city"}}, "required": ["city"]}
}

set_weather_tool = {
    "type": "function", "name": "set_weather", "description": "Sets the temperature for a specified city.",
    "parameters": {"type": "object", "properties": {"city": {"type": "string", "description": "The name of the city"}, "temp": {"type": "number", "description": "The temperature"}}, "required": ["city", "temp"]}
}

# --- 3. Use the chat_assistant framework ---
# Initialize the tools manager
tools = chat_assistant.Tools()
tools.add_tool(get_weather, get_weather_tool)
tools.add_tool(set_weather, set_weather_tool)

# Initialize the OpenAI client
client = OpenAI()

# Create a developer prompt to guide the agent
developer_prompt = """
You are a friendly weather assistant. Use the available tools to get or set the weather for a city.
If a city is not specified, you must ask the user for it.
""".strip()

# Initialize the chat interface and the main assistant
chat_interface = chat_assistant.ChatInterface()
chat = chat_assistant.ChatAssistant(
    tools=tools,
    developer_prompt=developer_prompt,
    chat_interface=chat_interface,
    client=client
)

# --- 4. Run the chat assistant ---
# Now you can interact with it in your terminal/notebook
# Try asking: "What's the weather in Berlin?"
# Then try: "Set the weather in Paris to 15 degrees"
# Then ask: "What's the weather in Paris?"
chat.run()
```
